```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
x=np.linspace(-5,5,1000)
def gaussian(x, mu, sigma):
    return multivariate_normal.pdf(x, mu, sigma)
```

```python
def true(x):
    return .7*gaussian(x, -.7, .4)+.3*gaussian(x, 1, 1)
```

```python
def proposal(x):
    return gaussian(x, 0, 2)
```

```python
x_sample = np.random.randn(10000)*2**.5 + 0
```

```python
plt.plot(x, true(x), label= 'posterior p', alpha=1, c='k', linewidth=2)
plt.ylim(0, .5)
#plt.plot(x, target, alpha=1, linewidth=3, label= 'proposal q')
plt.hist(x_sample, density=True, bins=100, histtype='step', label= 'proposal q', linewidth=3)
w=true(x_sample)/proposal(x_sample)
plt.hist(x_sample, density=True, bins=100, weights=w, align='mid',histtype='step', label= 'p from q', linewidth=3)
plt.xlabel('x', fontsize=20)
plt.xlim(-5, 5)
plt.legend(frameon=False, fontsize=12, loc='upper right')
plt.savefig('illustrate_is.pdf', bbox_inches='tight', dpi=300)
```
