```python
color = ['C0', 'C1', 'C2']
fig, ax = plt.subplots(1, 1, figsize=(5, 3), sharex=True)
file = file_hybrid
for i in range(9):
    
    #unbinned
    res_unbinned = plotssc.likelihood_load(file = file_unbinned[i])
    mean_Om_SN = np.trapz(res_unbinned['PSN']*res_unbinned['Om'], res_unbinned['Om'])
    err_Om_SN = np.trapz((mean_Om_SN - res_unbinned['Om'])**2*res_unbinned['PSN'], res_unbinned['Om'])**.5
    #plt.errorbar(N[i] - 0.05*N[i],mean_Om_SN, err_Om_SN,   marker = 'x', markersize=10,capsize=5, markerfacecolor='w', zorder=100)
    
    #hybrid
    res = plotssc.likelihood_load(file = file_hybrid[i])
    mean_Om_SN = np.trapz(res['PSN']*res['Om'], res['Om'])
    err_Om_SN = np.trapz((mean_Om_SN - res['Om'])**2*res['PSN'], res['Om'])**.5
    mean_Om_SSC = np.trapz(res['PSNSSC']*res['Om'], res['Om'])
    err_Om_SSC = np.trapz((mean_Om_SN - res['Om'])**2*res['PSNSSC'], res['Om'])**.5

    #plt.errorbar(N[i] + 0.0*N[i],mean_Om_SN,  err_Om_SN, marker = '+', markersize=10,capsize=5, markerfacecolor='w', color='k')
    plt.errorbar(N[i] + 0.05*N[i],mean_Om_SSC, err_Om_SSC,  marker = 'o', markersize=7,capsize=5, markerfacecolor='w', color='C0')
    
    #binned
    res_binned = plotssc.likelihood_load(file = file_binned[i])
    mean_Om_SN = np.trapz(res_binned['PSN']*res_binned['Om'], res_binned['Om'])
    err_Om_SN = np.trapz((mean_Om_SN - res_binned['Om'])**2*res_binned['PSN'], res_binned['Om'])**.5
    plt.errorbar(N[i] - 0.05*N[i],mean_Om_SN, err_Om_SN,   marker = 'x', markersize=10,capsize=5, markerfacecolor='w', zorder=100)
    
plt.errorbar([], [], marker = 'o', markersize=10, capsize=5, color = 'k', markerfacecolor='w', label = 'hybrid likelihood (SN+SSC)')
plt.errorbar([], [], marker = 'x', markersize=10, capsize=5, color = 'k', markerfacecolor='w', label = 'standard unbinned likelihood (SN-only)')
plt.errorbar([], [], marker = '+', markersize=10, capsize=5, color = 'k', markerfacecolor='w', label = 'hybrid likelihood (SN-only)')
plt.ylabel(r'$\Omega_m^{\rm fit}/\Omega_{m}^{\rm PINOCCHIO}$', fontsize=17)
plt.tick_params(axis='both', which = 'major', labelsize= 11)
plt.legend(fontsize=9, loc='upper right', frameon=True, framealpha=1)
plt.xscale('log')
#plt.savefig('cosmo.png', bbox_inches='tight', dpi=300)
```
