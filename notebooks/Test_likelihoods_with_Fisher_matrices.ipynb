```python
def dd_lnL(data):
    res=np.zeros([2,2])
    N=Abundance.flatten()
    for i in range(2):
        for j in range(2):
            res[i,j]=-np.sum(dd_Abundance[i,j]) + np.sum((data/N)*(
                (dd_Abundance[i,j].flatten()
                - d_Abundance[i].flatten()*d_Abundance[j].flatten()/N)))
    return res
```

```python
def d_lnLd_lnL(data):
    res=np.zeros([2,2])
    N=Abundance.flatten()
    for i in range(2):
        for j in range(2):
            di = -np.sum(d_Abundance[i])+np.sum(data*d_Abundance[i].flatten()/N)
            dj = -np.sum(d_Abundance[j])+np.sum(data*d_Abundance[j].flatten()/N)
            res[i,j]=di*dj
    return res
```

```python
def F(cov):
    res=np.zeros([2,2])
    N=Abundance.flatten()
    for i in range(2):
        for j in range(2):
            res[i,j]=np.sum((d_Abundance[i].flatten()/N)*cov.dot((d_Abundance[j].flatten()/N)))
    return res
```
